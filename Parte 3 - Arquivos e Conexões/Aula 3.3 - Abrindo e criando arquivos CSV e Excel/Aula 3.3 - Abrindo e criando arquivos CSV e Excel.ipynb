```python
# Importando as bibliotecas
import pandas as pd
import numpy as np
```

```python
# Criando um dataframe de exemplo
albuns = ["Bonito, Lindo e Joiado", 
"O Dinheiro Não é Tudo, Mas é 100%", 
"A Besteira é a Base da Sabedoria", 
"A Um Passo da MPB", 
"Quanto Pior, Melhor", 
"500 Anos de Chifre", 
"Do Penico à Bomba Atômica", 
"Maxximum", 
"What P***A iis This?", 
"Sucessão de Sucessos Que Se Sucedem Sucessivamente Sem Cessar",
"Under the Blade",
"You Can't Stop Rock'n Roll",
"Stay Hungry",
"Come Out and Play",
"Love is for Suckers",
"Big Hits and Nasty Cuts",
"Still Hungry"
]

anos = [
    1992, 1994, 1995, 1997, 1998, 1999, 2000, 2005, 2006, 2014,
    1982, 1983, 1984, 1985, 1987, 1992, 2004
]
artistas = []
for i in range(0,10):
    artistas.append("Falcão")
for i in range(0,7):
    artistas.append("Twisted Sister")

dados = {"Album" : albuns, "Ano": anos, "Artista": artistas}
discografia = pd.DataFrame(dados)
```

```python
discografia
```

```python
# Usando o to_csv para exportar o arquivo
discografia.to_csv("discografia.csv")
```

```python
# Usando o read_csv para ler o arquivo (usar o index_col)
a = pd.read_csv("discografia.csv", index_col=0)
```
